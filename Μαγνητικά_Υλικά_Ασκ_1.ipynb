```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import math
import decimal as dec
import sys

uo = 4 * np.pi / np.power(10,7)
data = pd.read_csv("data.csv")


H1 = data['H1'].to_numpy()
H1 = H1[~np.isnan(H1)]
B1 = data['B1'].to_numpy()
B1 = B1[~np.isnan(B1)]

H2 = data['H2'].to_numpy()
H2 = H2[~np.isnan(H2)]
B2 = data['B2'].to_numpy()
B2 = B2[~np.isnan(B2)]

H1above = data['H1above'].to_numpy()
H1above = H1above[~np.isnan(H1above)]
B1above = data['B1above'].to_numpy()
B1above = B1above[~np.isnan(B1above)]
H1below = data['H1below'].to_numpy()
H1below = H1below[~np.isnan(H1below)]
B1below = data['B1below'].to_numpy()
B1below = B1below[~np.isnan(B1below)]

H2above = data['H2above'].to_numpy()
H2above = H2above[~np.isnan(H2above)]
B2above = data['B2above'].to_numpy()
B2above = B2above[~np.isnan(B2above)]
H2below = data['H2below'].to_numpy()
H2below = H2below[~np.isnan(H2below)]
B2below = data['B2below'].to_numpy()
B2below = B2below[~np.isnan(B2below)]
```

```python

plt.figure(1)
plt.subplots(figsize = (25, 10))
plt.plot(H1,B1,'r')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel("B (T)")
plt.title("1st material B(H)  ")
plt.savefig("B-H_material1")
plt.show()
```

```python
plt.figure(2)
plt.subplots(figsize = (25, 10))
plt.plot(H2,B2,'g')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel("B (T)")
plt.title("2nd material B(H)")
plt.savefig("B-H_material2")
plt.show()
```

```python
M1 = np.empty(len(H1))
M2 = np.empty(len(H2))

for i in range(len(H1)):
    M1[i] = B1[i]/uo - H1[i]

for i in range(len(H2)):
    M2[i] = B2[i]/uo - H2[i]

plt.figure(3)
plt.subplots(figsize = (25, 10))
plt.plot(H1,M1,'r')
plt.grid()
plt.xlabel(" H (A/m)")
plt.ylabel(" M (A/m)")
plt.title("1st  material M(H)")
plt.savefig("M-H_material1")
plt.show()
```

```python
plt.figure(4)
plt.subplots(figsize = (25, 10))
plt.plot(H2,M2,'g')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel("M (A/m)")
plt.title("2nd material M(H)")
plt.savefig("M-H_material2")
plt.show()
```

```python
u1 = np.empty(len(H1))
u2 = np.empty(len(H2))

for i in range(len(H1)):
    u1[i] = B1[i]/H1[i]



for i in range(len(H2)):
    u2[i] = B2[i]/H2[i]

plt.figure(5)
plt.subplots(figsize = (25, 10))
plt.plot(H1,u1,'orange',label='below part')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel(" (H/m)")
plt.title("1st material (H)")
plt.savefig("u-H_material1")
plt.show()
```

```python
plt.figure(6)
plt.subplots(figsize = (25, 10))
plt.plot(H2,u2,'green',label='above part')
plt.grid()
plt.xlabel(" H (A/m)")
plt.ylabel(" (H/m)")
plt.title("2nd material (H)")
plt.savefig("u-H_material2")
plt.show()
```

```python
x1= np.empty(len(H1))
x2 = np.empty(len(H2))

for i in range(len(H1)):
    x1[i] = u1[i] / uo - 1

for i in range(len(H2)):
    x2[i] = u2[i] / uo - 1

plt.figure(7)
plt.subplots(figsize = (25, 10))
plt.plot(H1,x1,'red',label='above part')
plt.grid()
plt.xlabel("  (A/m)")
plt.ylabel("Magnetic susceptibility ")
plt.title("1st material (H)")
plt.savefig("x-H_material1")
plt.show()
```

```python
plt.figure(8)
plt.subplots(figsize = (25, 10))
plt.plot(H2,x2,'green',label='above part')
plt.grid()
plt.xlabel(" (A/m)")
plt.ylabel("Magnetic susceptibility ")
plt.title("2nd material (H)")
plt.savefig("x-H_material2")
plt.show()
```

```python
plt.figure(9)
plt.subplots(figsize = (25, 10))
plt.plot(H1, np.gradient(B1), 'red', label='above part')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel("d (H/m)")
plt.legend()
plt.title("1st material d(H)")
plt.savefig("ud1-H_material1")
plt.show()
```

```python
plt.figure(10)
plt.subplots(figsize = (25, 10))
plt.plot(H2, np.gradient(B2), 'green', label='above part')
plt.grid()
plt.xlabel("H (A/m)")
plt.ylabel(" d (H/m)")
plt.legend()
plt.title("2nd material d(H)")
plt.savefig("ud2-H_material1")
plt.show()
```

```python

print("----------------- 1st Material -----------------")
print("Negative coercive magnetic field is", H1above[pd.Series(abs(B1above)).idxmin()], "A/m")
print("Positive coercive magnetic field is", H1below[pd.Series(abs(B1below)).idxmin()], "A/m")

print("----------------- 2nd Material -----------------")
print("Negative coercive magnetic field is", H2above[pd.Series(abs(B2above)).idxmin()], "A/m")
print("Positive coercive magnetic field is", H2below[pd.Series(abs(B2below)).idxmin()], "A/m")



print("----------------- 1st Material -----------------")
print("Positive remaining magnetic induction is", B1above[pd.Series(abs(H1above)).idxmin()], "T")
print("Negative remaining magnetic induction is", B1below[pd.Series(abs(H1below)).idxmin()], "T")

print("----------------- 2nd Material-----------------")
print("Positive remaining magnetic induction is", B2above[pd.Series(abs(H2above)).idxmin()], "T")
print("Negative remaining magnetic induction is", B2below[pd.Series(abs(H2below)).idxmin()], "T")
```

```python
def FindArea(x,y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))

print("----------------- 1st Material-----------------")
print("Loop area = ", FindArea(H1,B1), "J")

print("----------------- 2nd Material-----------------")
print("Loop area = ", FindArea(H2,B2), "J")

```
